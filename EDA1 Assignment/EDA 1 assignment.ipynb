```python
median = df['LB'].median()
median
```

```python
#Filling missing value of LB with median value
df['LB']= df['LB'].fillna(median)
```

```python
df.isnull().sum()
```

```python
#All are continuous valueonly filling with median value 
median1 = df['AC'].median()
median2 = df['DS'].median()
median3 = df['DP'].median()
median4 = df['MLTV'].median()
median5 = df['Width'].median()
median6 = df['Tendency'].median()
median7 = df['NSP'].median()
df['AC']= df['AC'].fillna(median1)
df['DS']= df['DS'].fillna(median2)
df['DP']= df['DP'].fillna(median3)
df['MLTV']= df['MLTV'].fillna(median4)
df['Width']= df['Width'].fillna(median5)
df['Tendency']= df['Tendency'].fillna(median6)
df['NSP']= df['NSP'].fillna(median7)
```

```python
df.isnull().sum()
```

```python
#Oulier detecte for all variable 
import matplotlib.pyplot as plt

#Detect and treat outliers if necessary.
## Analyze the distribution
# Calculate skewness
skewness = df['LB'].skew()
print(f"Skewness: {skewness}")

# Calculate kurtosis
kurtosis = df['LB'].kurt()
print(f"Kurtosis: {kurtosis}")

# Describe data to identify potential outliers
description = df['LB'].describe()
print("Description:")
print(description)

# Calculate IQR to detect outliers
Q1 = df['LB'].quantile(0.25)
Q3 = df['LB'].quantile(0.75)
IQR = Q3 - Q1
print(f"IQR: {IQR}")

# Outlier detection
outliers = df[(df['LB'] < (Q1 - 1.5 * IQR)) | (df['LB'] > (Q3 + 1.5 * IQR))]
# Number of outliers
num_outliers = len(outliers)
print(f"Number of outliers: {num_outliers}")
```
